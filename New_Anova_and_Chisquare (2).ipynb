# Anova

```python
# Load libraries
from sklearn import datasets
# Load digits dataset
iris = datasets.load_iris()
```

```python
import pandas as pd
df = pd.DataFrame(iris.data)
df.head()
```

```python
import scipy.stats as stats
stats.f_oneway(df.iloc[:,0], df.iloc[:,1],df.iloc[:,2],df.iloc[:,3])
```

# **Chisquare**

```python
import pandas as pd
import numpy as np


df = pd.read_csv("mtcars.csv")
```

```python
df.head()
```

```python
crosstable = pd.crosstab(df['cyl'],df['gear'])
```

```python
crosstable
```

```python
from scipy.stats import chi2_contingency
chi, p, dof, expected = chi2_contingency(crosstable)
```

```python
print(chi,p,dof)
```
